### Steps of the text preprocessing

In order to process the text, we shall focus on the following aspects:

A) Finding punctuations

B) Removing punctuations

C) Removing null values

D) Word tokenization

E) Removing unwanted values

F) Removing special characters

### A) Finding possible punctuations

```python
import re
```

```python
def check_find_punctuations(text):
    
    result = re.findall(r'[!"\$%&\'()*+,\-.\/:;=#@?\[\\\]^_`{|}~]*', 
                        text)
    
    string= "".join(result)
    
    return list(string)


main_data['punctuation_used'] = main_data['text'].apply(lambda x : check_find_punctuations(x))
  
# show the Dataframe
main_data
```

Observation: We have filtered the possible punctuations based on the dataset

```python
import string
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[!"\$%&\'()*+,\-.\/:;=#@?\[\\\]^_`{|}~]*', '', text)
    return text
```

```python
main_data['final_text'] = main_data['text'].apply(lambda x: remove_punct(x))
main_data.head(10)
```
